```python
#Pandas program to compare the elements of two Pandas series.
#Sample Series [4,65,436,3,9], [7,0,3,897,9]

import pandas as pd

a = pd.Series([4,65,436,3,9])
b = pd.Series([7,0,3,897,9])

print("Original Series:")
print(a)
print(b)
print("\nCompare the elements of the said Series:")
print(a == b)

print("\nCheck whether elements of a series are greater than others:")
print(a > b)

print("\nCheck whether elements of a series are less than others:")
print(a < b)

print("\nCheck whether elements of a series are greater than or equal to others:")
print(a >= b)

print("\nCheck whether elements of a series are less than or equal to others:")
print(a <= b)

```

```python
# Pandas program to add, subtract, multiply and divide two Pandas series.
# Sample Series: [2,4,6,8,14], [1,3,5,7,9]

import pandas as pd

a = pd.Series([2,4,6,8,14])
b = pd.Series([1,3,5,7,9])

print("Original Series:")
print(a)
print(b)
print("\nAdd two series:")
print(a + b)

print("\nSubtract two series:")
print(a - b)

print("\nMultiply two series:")
print(a * b)

print("\nDivide two series:")
print(a / b)
```

```python
# Pandas program to display the most frequent value in the given series and replace
#  everything else as ‘Other’ in the series
import pandas as pd

# Load the dataset
file_path = 'AfricaCupofNationsMatches.csv'
df = pd.read_csv(file_path)

# Identify the most frequent value in the 'HomeTeamGoals' column
most_frequent_value = df['HomeTeamGoals'].mode()[0]
print(most_frequent_value)

# Replace all other values with 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: x if x == most_frequent_value else 'Other')

# Display the modified 'HomeTeamGoals' column to verify
print(df['HomeTeamGoals'])

```

```python
# Pandas program to reset index in your DataFrame
import pandas as pd

# Load the dataset
file_path = 'AfricaCupofNationsMatches.csv'
df = pd.read_csv(file_path)

# Retrieve the current index and sort it in descending order
sorted_index = df.index.sort_values(ascending=False)

# Re-index the DataFrame using the sorted index
df_reindexed = df.reindex(sorted_index)

print(df_reindexed.to_string())

print("\nAfter resetting")
# Reset index in your DataFrame
df_reset = df_reindexed.reset_index(drop=True)
print("\nDataFrame with Reset Index:\n", df_reset.head().to_string())
```
